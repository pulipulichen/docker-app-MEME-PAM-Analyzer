```python
import os
import pandas as pd
```

```python
def load_dataset():
    data = {'image': [], 'url': [], 'cluster': []}
    df = pd.DataFrame(data)
    
    # Directory to search for images in (replace with your directory)
    root_directory = 'input/data/'

    # List of image file extensions you want to search for
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']

    # Initialize a list to store the image file paths
    image_paths = []

    # Recursively search for image files in subdirectories
    for root, dirs, files in os.walk(root_directory):
        for file in files:
            if any(file.lower().endswith(ext) for ext in image_extensions):
                filepath = os.path.join(root, file)
                if filepath.endswith(".inpaint.jpg") is False and filepath.endswith(".crop.jpg") is False:
                    image_paths.append(filepath)

    # Print the list of image file paths
    # for image_path in image_paths:
    #     print(image_path)

    df['image'] = image_paths
    
    for index, row in df.iterrows():
        image_path = row['image']
        image_dir = os.path.dirname(image_path)

        url_file = os.path.join(image_dir, 'url.txt')
        cluster_file = os.path.join(image_dir, 'cluster.txt')

        if os.path.exists(url_file):
            with open(url_file, 'r') as file:
                df.at[index, 'url'] = file.read().strip()

        if os.path.exists(cluster_file):
            with open(cluster_file, 'r') as file:
                df.at[index, 'cluster'] = file.read().strip()
    
    
    nominal_list = df["cluster"].values.tolist()
    
    # Create a dictionary to map nominal items to index numbers
    item_to_index = {item: index for index, item in enumerate(set(nominal_list))}

    # Use a list comprehension to convert nominal items to index numbers
    index_list = [item_to_index[item] for item in nominal_list]

    df["cluster_index"] = index_list
    
    return df
```

```python
DATASET = load_dataset()
```
